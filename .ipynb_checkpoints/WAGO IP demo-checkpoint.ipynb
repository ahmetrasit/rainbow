### Read the file and prepare required fields

```python
df = pd.read_csv('wago1_enriched.genome.sam', header=None, sep='\t', usecols=range(5))
```

```python
df.columns = ['raw', 'flag', 'chrom', 'start', 'mapq']
```

```python
df['seq'] = df['raw'].map(lambda x: x.split(":")[0])
df['input'] = df['raw'].map(lambda x: float(x.split(":")[1]))
df['ip'] = df['raw'].map(lambda x: float(x.split(":")[2]))
```

```python
def getStrand(x):
    if x==0 or x==256:
        return '-'
    return '+'


df['strand'] = df['flag'].map(lambda x: getStrand(x))   
```

```python
df['len'] = df['seq'].map(lambda x: len(x))
df['end'] = df['start'] + df['len']
```

```python
df['type'] = df['seq'].map(lambda x: str(len(x))+x[0])
```

```python
#I need to calculate ratios, and there're zeros in input as well
df['input'] = df['input'].map(lambda x: 1 if x<1 else x)
df['ratio'] = df['ip']/df['input']
```

```python
df.columns
```

Re-organize columns and get rid of unnecessary ones

```python
df = df[['chrom', 'start', 'end', 'strand', 'mapq', 'seq', 'input', 'ip', 'ratio', 'type']]
```
